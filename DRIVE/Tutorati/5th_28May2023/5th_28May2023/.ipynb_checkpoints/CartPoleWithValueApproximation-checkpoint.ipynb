## Value-approximation

So far we dealt with value functions which were a perfect mapping of the states.
For each state (state/action pair), we had a tabular value of its value $V(s)$ (or $Q(s,a)$).
This is the __tabular__ approach.

To do so we learned the value function for _all_ states indipendently, which we have to visit a reasonably large number of times individually to get a good estimate!

Clearly this means that convergence scale badly if the number of states is very large, and it is impossible (with this simple formulation) if the state space is continuous.

To solve these problems most used in technological applications use what is called _value-approximation_.


## How does value-approximation work?

We need two ingredients:

1) How can I calculate the (approximate) Value given a state $s$.

2) How can I update the (approximate) Value given past experience such as a typical $(s,a,r',s',a')$ tuple.





#### 1. Calculation

We will define an auxiliary __weight__ vector $w$, such that the value is a function of both the state $s$ and $w$. We will write in a very general way $V(s, w)$, or $Q(s,a,w)$.
$w$ contains all our knowledge of the value function for all states. At a given time $t$, we will have one vector ${w}^*$: then, the value function for all states will be calculated as $V(s, w*)$, or $Q(s,a,w*)$.

#### 2. Update

Since the **weights** $w$ contain all information on the value, we will not update anymore the value $V$ or $Q$ itself, but **we will update the weights $w$ instead**.
The update will follow the rule. Where the values used to be changed with:

$$
V \leftarrow V + \alpha \delta \\
Q \leftarrow Q + \alpha \delta_Q \\
$$

Now at time $t$ we update the weights as:

$$
w \leftarrow w + \alpha \delta \nabla \hat{V}(S_t, \mathbf{w}) \\
w \leftarrow w + \alpha \delta_Q \nabla \hat{Q}(S_t, A_t, \mathbf{w})
$$

Where $\nabla f = (\frac{\partial f}{\partial w_1}, \frac{\partial f}{\partial w_2},\dots, \frac{\partial f}{\partial w_d} )$. So instead of changing $V$ directly as in previous cases, we update $V$ _through_ the change in the weights $w$.

### Today we are going to use cases of the two instances of _value-approximation_:


1) **Linear approximation**.
- *State aggregation*.
- *Polynomial features*.

2) **Non-linear approximation**.
- *Neural Network* (i.e. an universal function approximator).

### 1) Linear approximation.

First of all we construct a basis of $d$ _features_ $x = (x_1, x_2, \dots, x_d)$ to evaluate the value functions. 
Each state then has a feature vector of $d$ real-valued components ${x}(s) = (x_1(s), x_2(s), x_2(s), \dots, x_d(s))$.
The convenient part of the _features_ is that its dimension can be much smaller than the actual state space $d << |S|$.

Features can be defined in many ways (but clearly work best when they are chosen with care).
Then, the value function in the linear approximation reads as:

$$
V(s) = \sum_i^d w_i x_i(s) 
$$


_NB: The approximation can be done both for the state $s$ and for the action $a$! Since the "cart pole" example has only two allowed actions, we will deal only with approximation related to the state $s$ and keep the two actions distinct! So in our case we will have the Q-value in linear approximation, written as:_

$$
Q(s,a) = \sum_i^d w_i^a x_i(s) \,\,\,\,\,\,\,\, a=0,1
$$

And we will keep two sets of weights $\{w^a \}$, one for each actions. The method we use today is therefore _half_ tabular (in action space) and _half_ in function approximation (in state space).

### To recap: 
#### 1) Linear approximation means that the value is a linear combinations of the feature vector $x(s)$ of the state.
#### 2) How to choose the features is extremely relevant to how well the Reinforcement Learning will learn!

### Linear Approximation 1: State aggregation.

State aggregation is a very straightforward method to choose the features to reduce the size/dimension of the state space.
At its core, the state space is divided into regions $D_i$ and the components of $x(s)$ are just $x_i(s) = \mathbb{1}(s \in D_i)$.

This means that we still have a **tabular entry for values**, not for each state $s$ but **for each region $D_i$!**


### The implementation of the algorithm is VERY simple:

- In a state $S_t$ we find $x(S_t)$ and calculate $Q(S_t, a, w) = w^T x(S_t)$
- We use  $Q(S_t, a, w)$ to select an action  $A_t$
- Acting on state $S_t$ with action $A_t$ we obtain the reward $R_{t+1}$ and the new state $S_{t+1}$
- We map $S_{t+1}$ in the corresponding $x(S_{t+1})$ (and $S_t$ to $x(S_{t})$)
- We update the value of the weights $w$ using $R_{t+1}$ and the previously estimated $Q(S_t, A_t, w)$


What happens at the term $\nabla \hat{Q}(S_t, A_t, \mathbf{w})$? Let us see what the estimate $Q$ is in terms of the features $x$ and weights $w$:

$$
Q(S_t, A_t, \mathbf{w}) =  \sum_i \, w_i^{a=A_t} \, x_i(S_t) \\
Q(S_t, A_t, \mathbf{w}) =  \sum_i \, w_i^{a=A_t} \, \mathbb{1}(S_t \in D_i)\\
Q(S_t, A_t, \mathbf{w}) = w_j^{a=A_t}  
$$

Where $j$ is exactly the region to which $S_t$ belongs. 
In practice, the weights are exactly the estimation of the value for that region.
($\nabla \hat{Q}(S_t, A_t, \mathbf{w}) = (0,0,\dots,1,\dots,0)$, i.e. a vector of zeros with only a single $1$ at position $j_{th}$.)

#### -----------------------------

One of the simplest way to create the regions is through _binnings_. Essentially, a continuous variable $s$ defined in the interval $[s_{start}, s_{end}]$ could be divided into $N$ bins $\{s_i\}$, such that $x(s) = s_i$ if:

$$
s_{start} + i \cdot (s_{end} + s_{start})/N < s < s_{start} + (i+1)\cdot (s_{end} + s_{start})/N 
$$. 

We can do this in many dimensions. For example in 2D, we can divide each dimension, than for each state we have two identifying integers.





State aggregation can be done in many ways. If for example one property of the state is considered not relevant for the task, one could create regions that aggregate _all_ different values of that property into the same feature.




Clearly, the approximation can be as good as we wish, at the (heavy) price of having very small regions.


### FORMALLY...

Let us say we have $N_1$ and $N_2$ different bins in the two dimension. 
The feature vectors $x(s)$ are actually 1D vectors of lenght $N=N_1 * N_2$ where only $1$ entry is non-zero (for example, index $k = i+N_1*j$)

$$
x(s) = (0,0,0,0,0,\dots,1,\dots,0,0,0,0,0)
$$

With two dimensions, it is perhaps simpler to visualize them as a 2D matrix... But $x(s)$ is truly a 1D vector of projections $(x_1(s), x_2(s), \dots, x_{N_1*N_2}(s))$

$$
\begin{bmatrix}
    0       & 0 & \dots & 0 & \dots & 0 \\
    0       & 0 & \dots & 0 & \dots & 0 \\
    \vdots & \vdots & \ddots & \vdots&\vdots &0\\
    0 & 0& \dots & 1 & \dots & 0  \\
    \vdots & \vdots & \vdots & \vdots & \ddots &0 \\
    0      & 0 &  0 & 0 & 0 &0
\end{bmatrix}
$$

The weights vector has the same dimension, and contains the current estimation for the value function:

$$
w = (V_0,V_1,\dots,V_d)
$$

or
$$
\begin{bmatrix}
    V_{0 0}  \dots &  \dots & V_{0 j} &\dots & V_{0 N_2} \\
    \vdots &  \ddots & \vdots&\vdots & \vdots\\
    V_{i 0} & \dots & V{i j} & \dots & V_{i N_2}  \\
    \vdots & \vdots & \vdots &  \ddots & \vdots \\
    V_{N_1 0}     & \dots & V_{N_1 j} & \dots & V_{N_1 N_2} 
\end{bmatrix}
$$

Each operation $ Q = w^T x$ is done multiplying each component of $x$ with each component of $w$... But clearly of the whole sum only ONE term is non-zero! The one relative to the bin of the state $s$!

## Practically...
But in the practical use of it, we need only the index corresponding with $x(s)$, and use it to find the corresponding value in the vector $w$.



In the case of state aggregation **we would not even need to change the algorithm**, we could do the binning (i.e. feature extraction) outside the algorithm and provide to the algorithm the $x(S_t)$ instead of $S_t$!

But for clarity sake, let us implement it in our linear approximation version of SARSA.

- We need to compute $Q$ with the linear approximation everywhere we used $Q$ in the last lecture (in the update, and in the $\epsilon$-greedy function).

```python
class SARSA_TDControl_state_aggregation():
    def __init__(self, 
                 space_size, 
                 action_size, 
                 binning,
                 Nbinning,
                 gamma=1, 
                 lr_v=0.01):
        """
        Calculates optimal policy using in-policy Temporal Difference control
        Approximate Q-value for (S,A) pairs, using one-step updates and state aggregation!
        """        
        # the discount factor
        self.gamma = gamma
        # size of system
        # We don't truly need the space_size anymore, we work in feature space!
        self.feature_size = space_size
        
        # action size
        self.action_size = action_size

        # needed for state aggregation
        self.binning = binning
        self.Nbinning = Nbinning
        
        # the learning rate
        self.lr_v = lr_v
        
        # We used to have Qvalues... now we have weights!
        # self.Qvalues = np.zeros( (*self.state_space, self.action_size) )
        Nbins = self.Nbinning
        
        self.w = np.zeros((*Nbins, self.action_size))+50
        
    def state_aggregate(self, state):
        # Nbinning is a tuple (N1, N2, N3, N4), with #bins for each of the four dimensions
        # which define dimension of features
        Nbins = self.Nbinning
        x = np.zeros((*Nbins,))
        
        # x_indexes is the (only) index equal to 1 as dictated by the formula 1(s in D_s)
        x_indexes = np.array([0,0,0,0], dtype='int')
        
        for i,s in enumerate(state):
            if s <= self.binning[i,0]:
                x_indexes[i] = 0
            elif s >= self.binning[i,1]:
                x_indexes[i] = self.Nbinning[i]-1
            else :
                x_indexes[i] = np.floor((s-self.binning[i,0])/(self.binning[i,1]-self.binning[i,0])*self.Nbinning[i])
                
        x[(*x_indexes,)] = 1
        return x, x_indexes
        # Now x is truly a feature vector (in this case matrix for convenience) of all zeros except for one 1!
        # X_indexes is the "shorthand", that is the exact position of the x
        
    # -------------------   
    def single_step_update(self, s, a, r, new_s, new_a, done):
        """
        Uses a single step to update the values, using Temporal Difference for Q values.
        Employs the EXPERIENCED action in the new state  <- Q(S_new, A_new).
        """
        
        x, x_index = self.state_aggregate(s)
        new_x, new_x_index = self.state_aggregate(new_s)
        
        if done:
            # -----------------------
            # --- in SARSA it was ---
            # deltaQ = (r + 0 - self.Qvalues[ (*s, a) ])
            
            # -----------------------
            # --- Formally, we should use:
            # deltaQ = (r + 0 - np.tensordot(self.w[:,:,:,:, a], x, axes=4))
            # -----------------------
            
            # -----------------------
            # but since only one term is non-zero it is sufficient to use: 
            deltaQ = (r + 0 - self.w[(*x_index, a)])
            
        else:
            # --------------------------
            # in SARSA it was
            # deltaQ = (r + 
            #          self.gamma * self.Qvalues[ (*new_s, new_a) ] 
            #                     - self.Qvalues[ (*    s,     a) ])
            # ---------------------------
                      
            # -----------------------
            # --- Formally, we should use:
            # deltaQ = (r + 
            #          self.gamma * np.tensordot(self.w[:,:,:,:, new_a], new_x, axes=4) 
            #                     - np.tensordot(self.w[:,:,:,:, a], x, axes=4))
            
            # ... but it would be sufficient to use
            deltaQ = (r + 
                      self.gamma * self.w[(*new_x_index, new_a)]
                                 - self.w[(*x_index, a)])
            
        # --------------------
        # in SARSA it was
        # self.Qvalues[ (*s, a) ] += self.lr_v * deltaQ
        # --------------------
                      
        # --------------------
        # Formally, it is a VECTOR update with x (even if now x is just all zeros except for one...)
        # self.w[:,:,:,:, a] += self.lr_v * deltaQ * x
        # --------------------
        
        # --------------------
        # But since it is just one non-zero term
        # it is sufficient to use:
        self.w[(*x_index, a)] += self.lr_v * deltaQ
        
    # ---------------------
    def get_action_epsilon_greedy(self, s, eps):
        """
        Chooses action at random using an epsilon-greedy policy wrt the current Q(s,a).
        """
        ran = np.random.rand()
                      
        # features of the state
        x, x_index = self.state_aggregate(s)    
        
        # Q value as defined by linear approximation
        # It is a vector with dimension equal to the action space:
        # Q (LEFT)    = \sum_i       w_i^L   x_i
        # Q (RIGHT)   = \sum_i       w_i^R   x_i
        # np.tensordot is just the sum over multiple axis: 
        # in this case we have four which correspond to the four binned dimensions.
        
        # So formally we should write
        # Qvalue_approx = np.tensordot(self.w, x, ((0,1,2,3), (0,1,2,3)))
                      
        # But it is sufficient to write:
        Qvalue_approx = self.w[(*x_index,)]
        
        if (ran < eps):
            # probability is uniform for all actions!
            prob_actions = np.ones(self.action_size) / self.action_size 
        else:
            # I find the best Qvalue
            best_value = np.max( Qvalue_approx )
            
            # There could be actions with equal value! 
            best_actions = ( Qvalue_approx == best_value )

            # best_actions is 
            # *True* if the value is equal to the best (possibly ties)
            # *False* if the action is suboptimal
            prob_actions = best_actions / np.sum(best_actions)
            
        # take one action from the array of actions with the probabilities as defined above.
        a = np.random.choice(self.action_size, p=prob_actions)
        return a 
        
    def greedy_policy(self, s):
        x, x_index = self.state_aggregate(s)  
                      
        # So formally we should write
        # Qvalue_approx = np.tensordot(self.w, x, ((0,1,2,3), (0,1,2,3)))
                      
        # But it is sufficient to write:
        Qvalue_approx = self.w[(*x_index,)]
        greedy_pol = np.argmax(Qvalue_approx)
        return greedy_pol, Qvalue_approx
```

```python
# TD(0) CONTROL: SARSA over CartPole with state aggregation.

n_episodes = 2000

gamma = 1
binning = np.array([[-0.5,0.5],[-0.5,0.5],[-0.09,0.09],[-0.5,0.5]])
Nbinning=(1,1,3,2)


# Initialize 
epsilon_0 = 1
lr_v_0 = 1

# Initialize algorithm
SARSA = SARSA_TDControl_state_aggregation(space_size=4, 
                                          action_size=2, 
                                          binning=binning,
                                          Nbinning=Nbinning,
                                          gamma=gamma, 
                                          lr_v=lr_v_0)


performance_traj_stateAggr = np.zeros(n_episodes)

count = 0
tstar = 0
epsilon = epsilon_0

# RUN OVER EPISODES
for i in range(n_episodes):
    done = False

    if (i%10==0):
        print(i, end=' ')
    s = env.reset()
    a = SARSA.get_action_epsilon_greedy(s, epsilon)
    
    while not done:


        # keeping track for convergence
        count += 1
        
        # Evolve one step
        new_s, r, done, info = env.step(a)
        
        # Keeps track of performance for each episode
        performance_traj_stateAggr[i] += r
        # Choose new action index
        new_a = SARSA.get_action_epsilon_greedy(new_s, epsilon)
        #print(s,act,a, r,new_s,new_a, done, ' Qvalue ', SARSA.Qvalues[(*s,)])
        # Single update with (S, A, R', S', A')
        SARSA.single_step_update(s, a, r, new_s, new_a, done)
        
        a = new_a
        s = new_s
        
        if count > tstar:
            # UPDATE OF LEARNING
            SARSA.lr_v = lr_v_0/(1 + 0.001*(count - tstar)**0.75)
            # UPDATE OF EPSILON
            epsilon = epsilon_0/(1. + 0.001*(count - tstar)**1.05)
        
    #print(SARSA.greedy_policy([0,0,0.2,0.1]),  SARSA.greedy_policy([0,0,0.2,-0.1]), SARSA.greedy_policy([0,0,0.4,0.1]), SARSA.greedy_policy([0,0,0.4,-0.1]))
        
        
print(count)
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(np.arange(n_episodes), performance_traj_stateAggr, label='SARSA state aggregation')
ax.legend()
```

### Linear Approximation 2: Polynomial features.


Polynomial features are a different method to reduce the size/dimension of the state space. 
The features are constructed over the state $s$ using a polynomial expansion. 

$$
x(s) = x_0 + A_0 s + s^T A_1 s + \dots \\ 
$$

Or in other words, $x$ is a vector made by components:

$$
x_i(s) = x_{i,0} + \sum_j A_{ij} s_j + \sum_{jk} A_{ijk} s_j s_k + \dots
$$

In practice, $x$ is a polynomial over the state. One very simple case, is that 

$$
x_i(s) = s_i
$$

In the cart-pole case, for example, we indeed have that each state $s$ corresponds to _four_ real numbers: 
- $s_1 =$ Position of the cart along the x-axis. 
- $s_2 =$ Cart velocity. 
- $s_3 =$ Pole angle. 
- $s_4 =$ Pole velocity at its tip

Then, our features $x$ correspond to the state.
The linear approximation then consitutes in the fact that we **approximate the value function with a linear relationship** between the state and some parameter vector $w$, so that for example $V(s | w) = w^T x(s)$. 

This is limiting, of course, and in most case it may fail miserably.
Cart-pole (which can be thought of as an inverted pendulum) is one of the cases in which it fails miserably :)

Once the feature calculation is done, however, the control algorithm is **exactly the same as above.** 

```python
class SARSA_TDControl_linear_features():
    def __init__(self, 
                 space_size, 
                 action_size, 
                 gamma=1, 
                 lr_v=0.01):
        """
        Calculates optimal policy using in-policy Temporal Difference control
        Approximate Q-value for (S,A) pairs, using one-step updates and polynomial features(first order)!
        """        
        # the discount factor
        self.gamma = gamma
        # size of system
        # We don't truly need the space_size anymore, we work in feature space!
        self.feature_size = space_size
        
        # action size
        self.action_size = action_size

        # needed for state aggregation
        self.binning = binning,
        self.Nbinning = Nbinning,
        
        # the learning rate
        self.lr_v = lr_v
        
        # We used to have Qvalues... now we have weights!
        # self.Qvalues = np.zeros( (*self.state_space, self.action_size) )
        self.w = np.zeros((self.feature_size, self.action_size))
        
    def polynomial_features(self, state):
        
        # first choice is:
        # (x1, x2, x3, x4) = (s1, s2, s3, s4)
        # very simple!
        x = state
        
        # we could also go to the full second order, if we change size of w (feature_size)
        # s = state
        # x = [1, 
        #      s[0], s[1], s[2], s[3], 
        #      s[0]*s[0], s[0]*s[1], s[0]*s[2], s[0]*s[3], s[1]*s[1], s[1]*s[2], s[1]*s[3], s[2]*s[2], s[2]*s[3], s[3]*s[3]] 
        
        
        return x
        # Now x is a vector of real-values of dimension 4 (15 in the full second order) 
        
    # -------------------   
    def single_step_update(self, s, a, r, new_s, new_a, done):
        """
        Uses a single step to update the values, using Temporal Difference for Q values.
        Employs the EXPERIENCED action in the new state  <- Q(S_new, A_new).
        """
        
        x = self.polynomial_features(s)
        new_x = self.polynomial_features(new_s)
        
        if done:
            # in SARSA it was
            # deltaQ = (r + 0 - self.Qvalues[ (*s, a) ])
            deltaQ = (r + 0 - np.dot(self.w[:, a], x))
        else:
            # in SARSA it was
            # deltaQ = (r + 
            #          self.gamma * self.Qvalues[ (*new_s, new_a) ] 
            #                     - self.Qvalues[ (*    s,     a) ])
            deltaQ = (r + 
                      self.gamma * np.dot(self.w[:, new_a], new_x) 
                                 - np.dot(self.w[:, a], x))
            
        # in SARSA it was
        # self.Qvalues[ (*s, a) ] += self.lr_v * deltaQ
        # 
        # Now it is a VECTOR update with x (even if now x is just all zeros except for one...)
        self.w[:, a] += self.lr_v * deltaQ * x
        
    # ---------------------
    def get_action_epsilon_greedy(self, s, eps):
        """
        Chooses action at random using an epsilon-greedy policy wrt the current Q(s,a).
        """
        ran = np.random.rand()
                      
        # features of the state
        x = self.polynomial_features(s)    
        
        # Q value as defined by linear approximation
        # It is a vector with dimension equal to the action space:
        # Q (LEFT)    = \sum_i       w_i^L   x_i
        # Q (RIGHT)   = \sum_i       w_i^R   x_i
        Qvalue_approx = np.tensordot(self.w, x, ((0), (0)))
        
        if (ran < eps):
            # probability is uniform for all actions!
            prob_actions = np.ones(self.action_size) / self.action_size 
        
        else:
            # I find the best Qvalue
            best_value = np.max( Qvalue_approx )
            
            # There could be actions with equal value! 
            best_actions = ( Qvalue_approx == best_value )


            # best_actions is 
            # *True* if the value is equal to the best (possibly ties)
            # *False* if the action is suboptimal
            prob_actions = best_actions / np.sum(best_actions)
            
        # take one action from the array of actions with the probabilities as defined above.
        a = np.random.choice(self.action_size, p=prob_actions)
        return a 
        
    def greedy_policy(self, s):
        x, _ = self.state_aggregate(s)  
        Qvalue_approx = np.tensordot(self.w, x, ((0), (0))) 
        greedy_pol = np.argmax(Qvalue_approx)
        return greedy_pol
```

```python
# TD(0) CONTROL: SARSA over CartPole with polynomial features(first order) and linear appoximation.

n_episodes = 2000

gamma = 1

# Initialize 
epsilon_0 = 1
lr_v_0 = 1

# Initialize 
SARSA = SARSA_TDControl_linear_features(space_size=4, 
                                          action_size=2, 
                                          gamma=gamma, 
                                          lr_v=lr_v_0)

performance_traj_linear_features = np.zeros(n_episodes)

count = 0
tstar = 0
epsilon = epsilon_0

# RUN OVER EPISODES
for i in range(n_episodes):
    done = False

    s = env.reset()
    a = SARSA.get_action_epsilon_greedy(s, epsilon)
    
    if (i%10 == 0):
        print(i, end=' ')
    while not done:
    #for _ in range(100):
    
        # for convergence
        count += 1
        # Evolve one step
        new_s, r, done, info = env.step(a)

        # Keeps track of performance for each episode
        performance_traj_linear_features[i] += r
        # Choose new action index
        new_a = SARSA.get_action_epsilon_greedy(new_s, epsilon)
        #print(s,act,a, r,new_s,new_a, done, ' Qvalue ', SARSA.Qvalues[(*s,)])
        # Single update with (S, A, R', S', A')
        SARSA.single_step_update(s, a, r, new_s, new_a, done)
        
        a = new_a
        s = new_s

        if count > tstar:
            # UPDATE OF LEARNING
            SARSA.lr_v = lr_v_0/(1 + 0.001*(count - tstar)**0.75)
            # UPDATE OF EPSILON
            epsilon = epsilon_0/(1. + 0.001*(count - tstar)**1.05)        
        
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(np.arange(n_episodes), performance_traj_linear_features, label='SARSA linear features')
ax.legend()
```

### Beyond Linear Approximation: Neural Networks.


Linear Approximation (i.e. Value functions are linear superpositions of the features $x$ of the state) is a robust method but its results depend heavily on the way the features are constructed.

Poor choices of features (as for example $x = s$) that do not capture the dynamics of the environments can fail miserably to produce good results.

Nowadays non-linear approximations are most common, especially using Neural Networks.

At their core, Neural Networks are _universal function approximators_, i.e. they aim to reproduce any given function $f(x)$ using $f_{NN}(x | w)$, a fitting function with a very large of internal parameters $w$, some particular architecture and some overall common features of non-linearity.

**Why fitting?**
Remember that we are trying to _minimize the Value Error_.

$$
\bar{VE}(w) = \mathbb{E}_\pi\bigg[ \big(V_{\pi}(S_t) - \hat{v}(S_t,w)\big)^2 \bigg]
$$

So essentially we use the NN as a fitting function, trying to minimize the distance between its output - $\hat{v}(S,w)=f_{NN}{(S|w)}$ - and the target value.

#### Neural Network's basics

Generally speaking they have the structure of a networks of nodes and edges that start from an input ($x$) and ends up to few terminal nodes $f_{NN}(x | w)$, where the nodes correspond to fixed _operations_ (summation, products, filtering...) and the edges to variable parameters $w$.

Here we will deal with the most basic construction of Neural Networks where the structure will consist of layers of nodes one after the other, and the operation will be linear operations combined with non-linear filters.

So a Neural Network is a collection of possibly millions of weights and thousands of individual operations... Why should it be convenient to use?


**How do we get $\nabla V(s,a | w)$ if $w$ is a collection of millions of weights?**

Using the chain-rule of differentiation and the back-propagation of the gradients.

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

class SARSA_TDControl_NeuralNetwork():
    def __init__(self, 
                 space_size, 
                 action_size, 
                 gamma=1, 
                 lr_v=0.01):
        """
        Calculates optimal policy using in-policy Temporal Difference control
        Evaluates Q-value for (S,A) pairs, using one-step updates.
        """        
        # the discount factor
        self.gamma = gamma
        # size of system
        # We don't truly need the space_size anymore, we work in feature space!
        self.input_size = space_size
        
        # action size
        self.action_size = action_size
        
        # the learning rate
        self.lr_v = lr_v
        our_SGD = keras.optimizers.SGD(learning_rate=lr_v)
        
        # We used to have Qvalues... now we have one Neural Network for each action!
        # Model is as such:
        
        self.Q_NN_approx = []
        
        for a in range(action_size):
            # Definition of NN for approximate Q of left action
            NN = Sequential()
            NN.add(Dense(20, input_dim=self.input_size, activation='relu'))
            NN.add(Dense(20, activation='relu'))
            NN.add(Dense(1, activation='linear'))
            # "Construct" the NN with all the internal structure needed for calculations and updates..
            NN.compile(loss='MSE', optimizer=our_SGD)
            self.Q_NN_approx.append(NN)

        # At the end self.Q_NN_approx is a list with NNs for each action!
        
                
    # -------------------   
    def single_step_update(self, s, a, r, new_s, new_a, done):
        """
        Uses a single step to update the values, using Temporal Difference for Q values.
        Employs the EXPERIENCED action in the new state  <- Q(S_new, A_new).
        """
        
        # Just need a different shape for the Neural Network
        x = s.reshape(1,self.input_size)
        new_x = new_s.reshape(1,self.input_size)
        
        # I apply the Neural Network to get the current estimation of the Q(s,a) and Q(new_s,new_a)
        Q_approx = self.Q_NN_approx[a](x)
        new_Q_approx = self.Q_NN_approx[new_a](new_x)
        
        # Now Q_approx (and new_Q_approx) are value for the pair (s,a) (and (new_s, new_a))
        
        if done:
            # in SARSA it was
            # deltaQ = (r + 0 - self.Qvalues[ (*s, a) ])
            # deltaQ = (r + 0 - Q_approx)
            target = r + 0
        else:
            # in SARSA it was
            # deltaQ = (r + 
            #          self.gamma * self.Qvalues[ (*new_s, new_a) ] 
            #                     - self.Qvalues[ (*    s,     a) ])
            # deltaQ = (r + 
            #          self.gamma * new_Q_approx
            #                     -     Q_approx)
            target = r + self.gamma * new_Q_approx
            
        # in SARSA it was
        # self.Qvalues[ (*s, a) ] += self.lr_v * deltaQ
        # 
        # in Linear Approximation it was 
        # w <- deltaQ(s,a|w) nabla Q(s,a|w)

        target = np.array(target).reshape(1,1)
        
        # Here the update with all the gradients is done via the Keras library
        self.Q_NN_approx[a].fit(x, target, verbose=0)

        
    # ---------------------
    def get_action_epsilon_greedy(self, s, eps):
        """
        Chooses action at random using an epsilon-greedy policy wrt the current Q(s,a).
        """
        ran = np.random.rand()
        
        s = s.reshape(1,self.input_size)
        # Q value as defined by Neural Networks
        # It is a vector with dimension equal to the action space:
        Qvalue_approx = np.array([self.Q_NN_approx[a](s) for a in range(self.action_size)]).flatten()
        
        
        if (ran < eps):
            # probability is uniform for all actions!
            prob_actions = np.ones(self.action_size) / self.action_size 
        
        else:
            # I find the best Qvalue
            best_value = np.max( Qvalue_approx )
            
            # There could be actions with equal value! 
            best_actions = ( Qvalue_approx == best_value )


            # best_actions is 
            # *True* if the value is equal to the best (possibly ties)
            # *False* if the action is suboptimal
            prob_actions = best_actions / np.sum(best_actions)
            
        # take one action from the array of actions with the probabilities as defined above.
        a = np.random.choice(self.action_size, p=prob_actions)
        return a 
        
    def greedy_policy(self, s):
        Qvalue_approx = np.array([self.Q_NN_approx[a](s) for a in range(action_space)])
        greedy_pol = np.argmax(Qvalue_approx)
        return greedy_pol
```
